Remove columns that do not contribute to predicting employee attrition, such as unique identifiers and constant-value columns.

# Task 4 – Model Building & Comparison

### Step 1: Split the Dataset into Training and Testing Sets

Split the preprocessed dataset into training (80%) and testing (20%) sets. The target variable is stratified to preserve the original class distribution.

```python
from sklearn.model_selection import train_test_split

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

print("Training Features :", X_train.shape)
print("Testing Features  :", X_test.shape)
print("Training Labels   :", y_train.shape)
print("Testing Labels    :", y_test.shape)
```

### Step 2: Train a Logistic Regression Model

Train a Logistic Regression classifier using balanced class weights to handle class imbalance and establish a baseline model.

```python
from sklearn.linear_model import LogisticRegression

# Initialize Logistic Regression model
log_model = LogisticRegression(
    class_weight="balanced",
    random_state=42,
    max_iter=1000
)

# Train the model
log_model.fit(X_train, y_train)

print("Logistic Regression model trained successfully.")
```

### Step 3: Train a Random Forest Classifier

Train a Random Forest classifier using balanced class weights for improved prediction on imbalanced data.

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight="balanced"
)

# Train the model
rf_model.fit(X_train, y_train)

print("Random Forest model trained successfully.")
```

### Step 4: Train a Gradient Boosting Classifier

Train a Gradient Boosting classifier to compare its performance with the other machine learning models.

```python
from sklearn.ensemble import GradientBoostingClassifier

# Initialize Gradient Boosting model
gb_model = GradientBoostingClassifier(
    random_state=42
)

# Train the model
gb_model.fit(X_train, y_train)

print("Gradient Boosting model trained successfully.")
```

### Step 5: Model Training Summary

Verify that all three machine learning models have been successfully trained.

```python
print("✔ Logistic Regression  : Trained")
print("✔ Random Forest        : Trained")
print("✔ Gradient Boosting    : Trained")
```

# Task 5 – Model Evaluation

### Step 1: Generate Predictions

Generate predictions on the test dataset using all three trained machine learning models.

```python
# Generate predictions for all models

log_pred = log_model.predict(X_test)
rf_pred = rf_model.predict(X_test)
gb_pred = gb_model.predict(X_test)
```

### Step 2: Evaluate Model Performance

Evaluate each model using Precision, Recall, F1-Score, and ROC-AUC Score.

```python
from sklearn.metrics import (
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

# Create an empty list
results = []

# Dictionary of models
models = {
    "Logistic Regression": (log_model, log_pred),
    "Random Forest": (rf_model, rf_pred),
    "Gradient Boosting": (gb_model, gb_pred)
}

# Evaluate each model
for name, (model, pred) in models.items():

    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    roc_auc = roc_auc_score(y_test, pred)

    results.append([
        name,
        precision,
        recall,
        f1,
        roc_auc
    ])

# Comparison table
results_df = pd.DataFrame(
    results,
    columns=[
        "Model",
        "Precision",
        "Recall",
        "F1-Score",
        "ROC-AUC"
    ]
)

results_df
```

### Step 3: Identify the Best Performing Model

Compare the evaluation metrics to determine which model provides the best overall performance for predicting employee attrition.

```python
# Display comparison table sorted by F1-Score

results_df = results_df.sort_values(
    by="F1-Score",
    ascending=False
).reset_index(drop=True)

results_df
```

### Observation

Among the three machine learning models, **Logistic Regression** achieved the best overall performance with the highest **Recall (0.617)**, **F1-Score (0.439)**, and **ROC-AUC Score (0.695)**. Since the objective of employee attrition prediction is to identify as many employees at risk of leaving as possible, Logistic Regression was selected as the best model for further evaluation.

### Step 4: Generate the Classification Report

Display the classification report of the best-performing model (Logistic Regression).

```python
from sklearn.metrics import classification_report

print(
    classification_report(
        y_test,
        log_pred
    )
)
```

### Step 5: Generate the Confusion Matrix

Create a confusion matrix to visualize the classification performance of the best-performing model.

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(
    y_test,
    log_pred
)

print(cm)
```

### Observation

The Logistic Regression model correctly identified **29 out of 47 employees** who were likely to leave the company, achieving a **Recall of 61.7%**. Although the model generated some false positive predictions, it successfully detected a majority of employees at risk of attrition. In an HR setting, identifying potential employee exits early is more valuable than missing employees who are likely to resign, making this model suitable for supporting retention strategies.

### Step 3: Plot the ROC Curve

Compare the ROC curves of all three machine learning models to evaluate their ability to distinguish between employees who stay and those who leave.

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Predicted probabilities
log_prob = log_model.predict_proba(X_test)[:, 1]
rf_prob = rf_model.predict_proba(X_test)[:, 1]
gb_prob = gb_model.predict_proba(X_test)[:, 1]

# ROC values
fpr_log, tpr_log, _ = roc_curve(y_test, log_prob)
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_prob)
fpr_gb, tpr_gb, _ = roc_curve(y_test, gb_prob)

# AUC values
auc_log = auc(fpr_log, tpr_log)
auc_rf = auc(fpr_rf, tpr_rf)
auc_gb = auc(fpr_gb, tpr_gb)

# Plot
plt.figure(figsize=(8,6))

plt.plot(fpr_log, tpr_log,
         label=f"Logistic Regression (AUC = {auc_log:.3f})")

plt.plot(fpr_rf, tpr_rf,
         label=f"Random Forest (AUC = {auc_rf:.3f})")

plt.plot(fpr_gb, tpr_gb,
         label=f"Gradient Boosting (AUC = {auc_gb:.3f})")

plt.plot([0,1],[0,1],'k--')

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")

plt.title(
    "ROC Curve Comparison of Classification Models",
    fontsize=15,
    fontweight="bold"
)

plt.legend()

plt.tight_layout()

plt.savefig(
    "charts/chart8_roc_curve.png",
    dpi=300
)

plt.show()
```

### Observation

The ROC curve comparison shows that **Logistic Regression** achieved the highest Area Under the Curve (AUC), indicating better overall discrimination between employees who are likely to leave and those who are likely to stay. Therefore, Logistic Regression was selected as the final model for business interpretation and HR recommendations.
